```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("SPY_close_price_5Y.csv", parse_dates=["Date"])

# Set Date as index (important for time-series)
df.set_index("Date", inplace=True)

# Plot Close price
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Close"], label="Close Price")

plt.title("SPY Close Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Add MA50 and MA200 in the dataframe
df["MA50"] = df["Close"].rolling(window = 54, min_periods=54).mean()
df["MA200"] = df["Close"].rolling(window = 365, min_periods=365).mean()

plt.figure(figsize=(14,7))
plt.plot(df.index, df["Close"], label="Close", alpha=0.8)
plt.plot(df.index, df["MA50"], label="50-day MA")
plt.plot(df.index, df["MA200"], label="200-day MA")

plt.yscale('log')
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("SPY with Moving Averages")
plt.legend()
plt.show()
```
